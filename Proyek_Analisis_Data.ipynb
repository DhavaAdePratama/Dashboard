## Import Semua Packages/Library yang Digunakan

````python
#Mengimpor semua library yang diibutuhkan
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

## Data Wrangling

### Gathering Data

````python
#input data dari day.csv
day_df = pd.read_csv('https://raw.githubusercontent.com/DhavaAdePratama/Dashboard/main/day.csv')
day_df.head()
````

````python
#input data dari hour.cvs
hour_df= pd.read_csv('https://raw.githubusercontent.com/DhavaAdePratama/Dashboard/main/hour.csv')
hour_df.head()
````

### Assessing Data

````python
#Mengecek missing value pada DF day
day_df.info()

day_df.isnull().sum()
````

````python
#Mengecek missing value pada DF_hour
hour_df.info()

hour_df.isnull().sum()
````

````python
print("jumlah duplikasi", day_df.duplicated().sum()) # melihat duplicated pada DF day
print("" ""
      "")
day_df.describe()
````

````python
print("jumlah duplikasi", hour_df.duplicated().sum()) # melihat duplicated pada DF hour
print("" ""
      "")
hour_df.describe()
````

### Cleaning Data

````python
#Mengubah tipe data pada dteday di day_df
datetime_columns = ['dteday']

for column in datetime_columns:
  day_df[column] = pd.to_datetime(day_df[column])

day_df.info()
````

## Exploratory Data Analysis (EDA)

### Explore ...

````python
#menggabungkan data pada day dan hour df dengan metode outer
day_hour_df = pd.merge(
    left=day_df,
    right=hour_df,
    how="outer",
    left_on="instant",
    right_on="instant"
)
day_hour_df.head()
````

````python
#mengelompokkan data berdasarkan musim
day_hour_df.groupby(by = 'season_x').nunique()
````

````python
#Menambahkan kolom bulan
day_hour_df['Month'] = day_hour_df['dteday_x'].dt.strftime('%B')

````

## Visualization & Explanatory Analysis

### Pertanyaan 1:

````python
#mmebuat group berdasarkan rata rata penyewaan perjam
monthly_df= day_hour_df.groupby(by = 'season_x').agg('hr').mean().reset_index()
monthly_df.head()


````

````python
#MEMBUAT VISUALISASI BY HOUR AVERAGE PER SEASON

plt.figure(figsize=(14, 7))
sns.barplot(x= 'season_x', y='hr', data=monthly_df, palette='coolwarm')
plt.xlabel('SEASON')
plt.ylabel('hr')
plt.title('HOUR AVERAGE')
plt.xticks(rotation=45)
plt.legend(title='Metric')
plt.show()
````

### Pertanyaan 2:

````python
#Mengganti angka pada season sesuai dengan musim yg tertera pada readme

day_hour_df.season_x.replace({
    1: 'SPRING',
    2: 'SUMMER',
    3: 'FALL',
    4: 'WINTER',
},inplace = True)
day_hour_df.head()
````

````python
# melakukan groupby untuk melihat berapa kali pemakaian dan jumlah jam pemakaian sepeda di setiap season
day_hour_df.groupby(by = 'season_x').agg({'hr' : ['count' ,'sum']})
````

````python
#membuat variabel penyimpan count
BY_COUNT = day_hour_df.groupby('season_x')['hr'].count().reset_index()

#membuat visualisasi
plt.figure(figsize=(10, 5))
plt.pie(
    BY_COUNT['hr'],
    labels=BY_COUNT['season_x'],
    autopct='%1.1f%%',
    startangle=140,
    colors=["#72BCD4", "#FFC0CB", "#FFB6C1", "#FF677D"]
)
plt.title("Total presentase pemakaian setiap musim", loc="center", fontsize=20)
plt.show()
````

````python
#membuat variabel penyimpan SUM
BY_SUM = day_hour_df.groupby('season_x')['hr'].sum().reset_index()

#membuat visualisasi
plt.figure(figsize=(10, 5))
plt.plot(
    BY_SUM["season_x"],
    BY_SUM['hr' ],
    marker='o',
    linewidth=2,
    color="#72BCD4"
)
plt.title("Jumlah jam pemakaian per musim", loc="center", fontsize=20)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()

````
