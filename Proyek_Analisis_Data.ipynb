## Import Semua Packages/Library yang Digunakan

````python
#Mengimpor semua library yang diibutuhkan
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

## Data Wrangling

### Gathering Data

````python
#input data dari day.csv
day_df = pd.read_csv('https://raw.githubusercontent.com/DhavaAdePratama/Dashboard/main/day.csv')
day_df.head()
````

````python
#input data dari hour.cvs
hour_df= pd.read_csv('https://raw.githubusercontent.com/DhavaAdePratama/Dashboard/main/hour.csv')
hour_df.head()
````

### Assessing Data

````python
#Mengecek missing value pada DF day
day_df.info()

day_df.isnull().sum()
````

````python
#Mengecek missing value pada DF_hour
hour_df.info()

hour_df.isnull().sum()
````

terdapat kesalahan type data pada kolom dteday

````python
print("jumlah duplikasi", day_df.duplicated().sum()) # melihat duplicated pada DF day
print("" ""
      "")
day_df.describe()
````

````python
print("jumlah duplikasi", hour_df.duplicated().sum()) # melihat duplicated pada DF hour
print("" ""
      "")
hour_df.describe()
````

### Cleaning Data

````python
#Mengubah tipe data pada dteday di day_df
datetime_columns = ['dteday']

for column in datetime_columns:
  day_df[column] = pd.to_datetime(day_df[column])

day_df.info()
````

MENGUBAH TYPE DATA PADA KOLOM dteday DARI object MENJADI datetime

## Exploratory Data Analysis (EDA)

### Explore ...

````python
#menggabungkan data pada day dan hour df dengan metode outer
day_hour_df = pd.merge(
    left=day_df,
    right=hour_df,
    how="inner",
    left_on="instant",
    right_on="instant"
)
day_hour_df.head()
````

pada proses ini kita menggabungkan data pada day dan hour df dengan menggunakan metode outer yang dimana semua data di day dan hour df akan digabungkan


````python
day_hour_df['weekday_avg'] = day_hour_df[['weekday_x', 'weekday_y']].mean(axis = 1)
day_hour_df.head()
````

menambahkan kolom rata rata pada weekday


````python
#mengelompokkan data berdasarkan musim
day_hour_df.groupby(by = 'season_x').nunique()
````

melihat jumlah data pada setiap musimnya


````python
#Menambahkan kolom bulan
day_hour_df['Month'] = day_hour_df['dteday_x'].dt.strftime('%B')

````

menambahkan kolom bulan

````python
day_hour_df.season_x.replace({
    1: 'SPRING',
    2: 'SUMMER',
    3: 'FALL',
    4: 'WINTER',
},inplace = True)
day_hour_df.head()
````

mengganti data numerik menjadi kategorik pada kolom season sesuai dengan keterangan di dataset.

## Visualization & Explanatory Analysis

### Pertanyaan 1:

````python
compare = day_hour_df[['weekday_avg', 'holiday_x']]
compare.head()
````

MEMBANDINGKAN DATA PADA KOLOM WEEKDAY DAN HOLIDAY_X

````python
count = day_hour_df.groupby('Month').agg({
    'weekday_avg' : 'sum',
    'holiday_x' : 'sum' }).reset_index()
count
````

MELIHAT TOTAL PENGGUNAAN DI SETIAP BULANNYA


````python
fig, ax = plt.subplots(figsize=(10, 6))

x = range(len(count))  # Posisi x untuk setiap bulan
width = 0.35  # Lebar bar

# Bar untuk weekday_avg
ax.bar(
    [p - width / 2 for p in x],
    count['weekday_avg'],
    width,
    label='Weekday Avg',
    color='skyblue'
)

# Bar untuk holiday_x
ax.bar(
    [p + width / 2 for p in x],
    count['holiday_x'],
    width,
    label='Holiday',
    color='orange'
)

# Menambahkan label dan keterangan
ax.set_xticks(x)
ax.set_xticklabels(count['Month'])
ax.set_xlabel('Month')
ax.set_ylabel('Sum of Values')
ax.set_title('Sum of Weekday Avg and Holiday by Month')
ax.legend()

# Menampilkan visualisasi
plt.tight_layout()
plt.show()
````

### Pertanyaan 2:

````python
# melakukan groupby untuk melihat berapa kali pemakaian dan jumlah jam pemakaian sepeda di setiap season
day_hour_df.groupby(by = 'season_x').agg({'hr' : ['count' ,'sum']})
````

MELIHAT JUMLAH JAM DAN BERAPA KALI PEMAKAIAN DI SETIAP MUSIMNYA  

````python
#membuat variabel penyimpan count
BY_COUNT = day_hour_df.groupby('season_x')['hr'].count().reset_index()

#membuat visualisasi
plt.figure(figsize=(10, 5))
plt.pie(
    BY_COUNT['hr'],
    labels=BY_COUNT['season_x'],
    autopct='%1.1f%%',
    startangle=140,
    colors=["#72BCD4", "#FFC0CB", "#FFB6C1", "#FF677D"]
)
plt.title("Total presentase pemakaian setiap musim", loc="center", fontsize=20)
plt.show()
````

````python
#membuat variabel penyimpan SUM
BY_SUM = day_hour_df.groupby('season_x')['hr'].sum().reset_index()

#membuat visualisasi
plt.figure(figsize=(10, 5))
plt.plot(
    BY_SUM["season_x"],
    BY_SUM['hr' ],
    marker='o',
    linewidth=2,
    color="#72BCD4"
)
plt.title("Jumlah jam pemakaian per musim", loc="center", fontsize=20)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()

````

hasil tersebut mengindikasikan bahwa pada fall dan summer di kedua musim tersebut penyewaan sepeda dalam tren yang positif sadangkan pada spring dan winter penyewaan sepeda menurun hasil tersebut dapat dilihat dari jumlah persentase pemakaian dan jumlah jam pemakaiaannya.
